subindo os dataframes pro MYSQL

obs.: foi necessário alterar o .info do SQL para aumentar o limite de trafego (fiz através do xampp)

```python
from sqlalchemy import create_engine
```

```python
conn = create_engine(
    'mysql+mysqlconnector://root:@localhost:3306/stream?', 
    pool_recycle=3600
)
```

```python
arquivo_games.to_sql(name="games", con=conn, if_exists='replace', index=False)
arquivo_global.to_sql(name="global", con=conn, if_exists='replace', index=False)
```
